# 01_intro figures

```python
import matplotlib.pyplot as plt
import numpy as np

# generate point cloud
np.random.seed(0)
x = np.linspace(0, 10, 30)
y = x + np.random.normal(scale=1.0, size=len(x))

# high leverage point
x_hl = np.append(x, 20)
y_hl = np.append(y, -10)

# Helper to save figure
def make_scatter(x_data, y_data, line=False, fname="fig.png",title=""):
    plt.figure()
    plt.scatter(x_data, y_data)
    if line:
        coef = np.polyfit(x_data, y_data, 1)
        xs = np.linspace(min(x_data), max(x_data), 100)
        ys = coef[0]*xs + coef[1]
        plt.plot(xs, ys,color='black')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.close()

# Create four figures
make_scatter(x, y, line=False, fname="00_introduction_pointcloud.png", title='Visual line fit')
make_scatter(x, y, line=True, fname="00_introduction_pointcloud_line.png", title='Visual line fit')
make_scatter(x_hl, y_hl, line=False, fname="00_introduction_pointcloud_high_leverage.png", title='Leverage fit')
make_scatter(x_hl, y_hl, line=True, fname="00_introduction_pointcloud_high_leverage_line.png", title='Leverage fit')

```

# 02_linear_regression figures

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# ------------------------------------------------------------------
# Settings
# ------------------------------------------------------------------
np.random.seed(1)



# ------------------------------------------------------------------
# Simulate data
# ------------------------------------------------------------------
n = 80

# Continuous predictor
x = np.random.uniform(0, 10, size=n)

# Binary categorical variable: sex (0/1)
sex = np.random.binomial(1, 0.5, size=n)

# True relationship:
#   y = beta0 + beta1 * x + beta_sex * sex + noise
beta0 = 2.0
beta1 = 0.8
beta_sex = 2.0

noise = np.random.normal(0, 1.0, size=n)
y = beta0 + beta1 * x + beta_sex * sex + noise

# ------------------------------------------------------------------
# Helper plotting functions
# ------------------------------------------------------------------
def save_simple_scatter(x, y, fname):
    plt.figure()
    plt.scatter(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Outcome vs. Predictor (Ignoring Sex)")
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.close()


def save_scatter_with_single_line(x, y, fname):
    plt.figure()
    plt.scatter(x, y)

    # Fit a single linear regression line ignoring sex
    coef = np.polyfit(x, y, 1)  # [slope, intercept]
    xs = np.linspace(x.min(), x.max(), 100)
    ys = coef[0] * xs + coef[1]

    plt.plot(xs, ys)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Single Linear Fit (Ignoring Sex)")
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.close()

def save_scatter_by_sex(x, y, sex, fname):
    plt.figure()

    # Mask for each sex
    mask0 = (sex == 0)
    mask1 = (sex == 1)

    # Scatter points with default colors for each group
    plt.scatter(x[mask0], y[mask0], label="Sex = 0", alpha=0.8)
    plt.scatter(x[mask1], y[mask1], label="Sex = 1", alpha=0.8)

 
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Sex separation")
    plt.legend()
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.close()

def save_scatter_by_sex_with_lines(x, y, sex, fname):
    plt.figure()

    # Mask for each sex
    mask0 = (sex == 0)
    mask1 = (sex == 1)

    # Scatter points with default colors for each group
    plt.scatter(x[mask0], y[mask0], label="Sex = 0", alpha=0.8)
    plt.scatter(x[mask1], y[mask1], label="Sex = 1", alpha=0.8)

    # Separate regression lines for each sex
    # Sex = 0
    coef0 = np.polyfit(x[mask0], y[mask0], 1)
    xs0 = np.linspace(x[mask0].min(), x[mask0].max(), 100)
    ys0 = coef0[0] * xs0 + coef0[1]
    plt.plot(xs0, ys0)

    # Sex = 1
    coef1 = np.polyfit(x[mask1], y[mask1], 1)
    xs1 = np.linspace(x[mask1].min(), x[mask1].max(), 100)
    ys1 = coef1[0] * xs1 + coef1[1]
    plt.plot(xs1, ys1)

    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Separate Fits by Sex")
    plt.legend()
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.close()


# ------------------------------------------------------------------
# Create and save the figures
#   (prefix: 01_linear_regression_*.png)
# ------------------------------------------------------------------
save_simple_scatter(
    x, y,
    fname="01_linear_regression_pointcloud_no_sex.png"
)

save_scatter_with_single_line(
    x, y,
    fname="01_linear_regression_pointcloud_single_line.png"
)

save_scatter_by_sex(
    x, y, sex,
    fname="01_linear_regression_scatter_by_sex.png"
)


save_scatter_by_sex_with_lines(
    x, y, sex,
    fname="01_linear_regression_by_sex_two_lines.png"
)
```

# 02_regularization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge

# Simulate correlated features
np.random.seed(42)
n, p = 80, 10
X = np.random.randn(n, p)
X = X + 0.5 * np.random.randn(n, 1)  # inject correlation
beta_true = np.array([3, -2, 0, 5] + [0]*(p-4))
y = X @ beta_true + np.random.randn(n) * 2

# Regularization strengths
alphas = np.logspace(-3, 2, 40)

ridge_paths = []
lasso_paths = []

for a in alphas:
    ridge = Ridge(alpha=a).fit(X, y)
    lasso = Lasso(alpha=a, max_iter=5000).fit(X, y)
    ridge_paths.append(ridge.coef_)
    lasso_paths.append(lasso.coef_)

ridge_paths = np.array(ridge_paths)
lasso_paths = np.array(lasso_paths)

def plot_paths(paths, alphas, fname, title):
    plt.figure()
    for i in range(paths.shape[1]):
        plt.plot(alphas, paths[:, i])
    plt.xscale("log")
    plt.xlabel("alpha")
    plt.ylabel("coefficient value")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.close()

path_ridge = "02_regularization_ridge_paths.png"
path_lasso = "02_regularization_lasso_paths.png"

plot_paths(ridge_paths, alphas, path_ridge, "Ridge (L2) Coefficient Shrinkage")
plot_paths(lasso_paths, alphas, path_lasso, "Lasso (L1) Coefficient Shrinkage")

path_ridge, path_lasso
```

# 03_non_linear

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# ---------- 1. Parabolic Example ----------
x1 = np.linspace(-5, 5, 80)
y1 = x1**2 + np.random.normal(scale=1.0, size=len(x1))

coef_lin1 = np.polyfit(x1, y1, 1)
xs = np.linspace(x1.min(), x1.max(), 200)
ys_lin1 = coef_lin1[0]*xs + coef_lin1[1]

coef_quad1 = np.polyfit(x1, y1, 2)
ys_quad1 = coef_quad1[0]*xs**2 + coef_quad1[1]*xs + coef_quad1[2]

# Figure 1
plt.figure()
plt.scatter(x1, y1)
plt.xlabel("x"); plt.ylabel("y")
plt.tight_layout()
plt.savefig("03_parabolic_pointcloud.png", dpi=150)
plt.close()

# Figure 2
plt.figure()
plt.scatter(x1, y1)
plt.plot(xs, ys_lin1, color='black')
plt.xlabel("x"); plt.ylabel("y")
plt.tight_layout()
plt.savefig("03_parabolic_linear_fit.png", dpi=150)
plt.close()

# Figure 3
plt.figure()
plt.scatter(x1, y1)
plt.plot(xs, ys_quad1, color='black')
plt.xlabel("x"); plt.ylabel("y")
plt.tight_layout()
plt.savefig("03_parabolic_quadratic_fit.png", dpi=150)
plt.close()


# ---------- 2. Dose–Response Example ----------
# S-shaped dose-response using a logistic function
x2 = np.linspace(0, 10, 80)
L = 5      
k = 3.0    
x0 = 5.0   
y2_clean = L / (1 + np.exp(-k * (x2 - x0)))
y2 = y2_clean + np.random.normal(scale=0.3, size=len(x2))

# Fits
coef_lin2 = np.polyfit(x2, y2, 1)
coef_quad2 = np.polyfit(x2, y2, 2)
coef_cubic2 = np.polyfit(x2, y2, 3)

xs2 = np.linspace(x2.min(), x2.max(), 200)
ys_lin2 = np.polyval(coef_lin2, xs2)
ys_quad2 = np.polyval(coef_quad2, xs2)
ys_cubic2 = np.polyval(coef_cubic2, xs2)



# Figure 4
plt.figure()
plt.scatter(x2, y2)
plt.xlabel("x"); plt.ylabel("y")
plt.tight_layout()
plt.savefig("03_dose_pointcloud.png", dpi=150)
plt.close()

# Figure 5
plt.figure()
plt.scatter(x2, y2)
plt.plot(xs2, ys_lin2, color='black')
plt.xlabel("x"); plt.ylabel("y")
plt.tight_layout()
plt.savefig("03_dose_linear_fit.png", dpi=150)
plt.close()

# Figure 6
plt.figure()
plt.scatter(x2, y2)
plt.plot(xs2, ys_quad2, color='black')
plt.xlabel("x"); plt.ylabel("y")
plt.tight_layout()
plt.savefig("03_dose_quadratic_fit.png", dpi=150)
plt.close()


# Cubic fit figure
plt.figure()
plt.scatter(x2, y2)
plt.plot(xs2, ys_cubic2, color='black')
plt.xlabel("x"); plt.ylabel("y")
plt.tight_layout()
plt.savefig("03_dose_cubic_fit.png", dpi=150)
plt.close()
```

# 04_mlp

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

# Common settings
np.random.seed(0)

# Helper to make meshgrid and decision boundary plots
def plot_decision_boundary(clf, X, y, fname, title=None):
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure()
    # contourf for background regions
    plt.contourf(xx, yy, Z, alpha=0.2)
    # scatter original points
    plt.scatter(X[:, 0], X[:, 1], c=y)
    if title:
        plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.close()


# =========================
# 1. Circular Classification
# =========================
n_samples = 400
radius = 1.6

# Generate points in a square, label by being inside/outside circle
X_circle = np.random.uniform(-2, 2, size=(n_samples, 2))
r = np.sqrt(X_circle[:, 0]**2 + X_circle[:, 1]**2)
y_circle = (r < radius).astype(int)

# Figure 1: raw circular data
plt.figure()
plt.scatter(X_circle[:, 0], X_circle[:, 1], c=y_circle)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Nonlinear Circular Classes")
plt.tight_layout()
plt.savefig("04_mlp_circle_points.png", dpi=150)
plt.close()

# Logistic regression
log_reg_circle = LogisticRegression().fit(X_circle, y_circle)

# Figure 2: logistic decision boundary
plot_decision_boundary(
    log_reg_circle, X_circle, y_circle,
    "04_mlp_circle_logistic_boundary.png",
    title="Logistic Regression (Linear Boundary)"
)

# MLP classifier
mlp_circle = MLPClassifier(hidden_layer_sizes=(20,), activation="tanh", max_iter=5000).fit(
    X_circle, y_circle
)

# Figure 3: MLP decision boundary
plot_decision_boundary(
    mlp_circle, X_circle, y_circle,
    "04_mlp_circle_mlp_boundary.png",
    title="MLP (Nonlinear Boundary)"
)


# =========================
# 2. XOR Pattern
# =========================
# Create XOR data in 2D
X_xor = np.random.randn(400, 2)
y_xor = (X_xor[:, 0] * X_xor[:, 1] > 0).astype(int)  # XOR pattern by sign

# Figure 4: raw XOR data
plt.figure()
plt.scatter(X_xor[:, 0], X_xor[:, 1], c=y_xor)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("XOR Pattern (Linearly Inseparable)")
plt.tight_layout()
plt.savefig("04_mlp_xor_points.png", dpi=150)
plt.close()

# Logistic regression on XOR
log_reg_xor = LogisticRegression().fit(X_xor, y_xor)

# Figure 5: XOR logistic decision boundary
plot_decision_boundary(
    log_reg_xor, X_xor, y_xor,
    "04_mlp_xor_logistic_boundary.png",
    title="Logistic Regression on XOR"
)

# MLP on XOR
mlp_xor = MLPClassifier(hidden_layer_sizes=(20,), activation="tanh", max_iter=5000).fit(
    X_xor, y_xor
)

# Figure 6: XOR MLP decision boundary
plot_decision_boundary(
    mlp_xor, X_xor, y_xor,
    "04_mlp_xor_mlp_boundary.png",
    title="MLP on XOR (Nonlinear Boundary)"
)
```
