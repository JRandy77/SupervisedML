# Regularization for High Dimensional Medical Data

Goal: Work with high dimensional medical like data and apply regularization.

```python
from workshop_utils import load_dataset, basic_train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

df = load_dataset("omics")
X_train, X_test, y_train, y_test = basic_train_test_split(df, target="inflammation_index")

# Baseline model
base_model = LinearRegression()
base_model.fit(X_train, y_train)
base_pred = base_model.predict(X_test)
print("Baseline MSE:", mean_squared_error(y_test, base_pred))

# TODO 2.1: set up Ridge with a parameter grid
# Tutorial: https://scikit-learn.org/stable/modules/linear_model.html#ridge-regression
ridge = Ridge()
ridge_grid = {"alpha": [0.1, 1.0, 10.0]}
ridge_cv = GridSearchCV(ridge, ridge_grid, cv=5)
ridge_cv.fit(X_train, y_train)
ridge_pred = ridge_cv.predict(X_test)
print("Ridge best alpha:", ridge_cv.best_params_)
print("Ridge MSE:", mean_squared_error(y_test, ridge_pred))

# TODO 2.2: do the same for Lasso
# Tutorial: https://scikit-learn.org/stable/modules/linear_model.html#lasso
lasso = Lasso(max_iter=5000)
lasso_grid = {"alpha": [0.001, 0.01, 0.1]}
lasso_cv = GridSearchCV(lasso, lasso_grid, cv=5)
lasso_cv.fit(X_train, y_train)
lasso_pred = lasso_cv.predict(X_test)
print("Lasso best alpha:", lasso_cv.best_params_)
print("Lasso MSE:", mean_squared_error(y_test, lasso_pred))

def train_regularized_model(X_train, y_train, model_type="ridge"):
    if model_type == "ridge":
        model = Ridge()
        grid = {"alpha": [0.1, 1.0, 10.0]}
    else:
        model = Lasso(max_iter=5000)
        grid = {"alpha": [0.001, 0.01, 0.1]}
    search = GridSearchCV(model, grid, cv=5)
    search.fit(X_train, y_train)
    return search
```
