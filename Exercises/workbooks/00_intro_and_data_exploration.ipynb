# Intro and Data Exploration

Goal: Load and explore a medical tabular dataset and create a train test split.

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

HERE = os.path.dirname(os.path.abspath(__file__))
DATA_DIR = os.path.join(HERE, "data")


def load_dataset(name: str) -> pd.DataFrame:
    base_url = (
        "https://raw.githubusercontent.com/"
        "Jesse-Islam/QLS-MiCM_Introduction_to_supervised_machine_learning/main/data/"
    )
    url = f"{base_url}{name}.csv"
    
    try:
        return pd.read_csv(url)
    except Exception as e:
        raise FileNotFoundError(f"Could not load dataset '{name}' from {url}") from e



def basic_train_test_split(df: pd.DataFrame, target: str, test_size: float = 0.2, random_state: int = 42):
    X = df.drop(columns=[target])
    y = df[target]
    return train_test_split(X, y, test_size=test_size, random_state=random_state)


def build_preprocessor(num_features, cat_features):
    numeric_transformer = StandardScaler()
    categorical_transformer = OneHotEncoder(handle_unknown="ignore")
    preprocessor = ColumnTransformer(
        transformers=[
            ("num", numeric_transformer, num_features),
            ("cat", categorical_transformer, cat_features),
        ]
    )
    return preprocessor
# Load and inspect
df = load_dataset("bp")
display(df.head())

# Split into train/test
X_train, X_test, y_train, y_test = basic_train_test_split(df, target="sbp")
print(X_train.shape, X_test.shape)
```

### Exploring Relationships Between Variables

Before we build any models, it is essential to **look** at the data.  
Visual exploration helps us:

- Detect trends or relationships (e.g., does SBP increase with age?).
- Spot outliers or unusual distributions.
- Develop hypotheses about which variables might matter most.

We'll use **Matplotlib** and **pandas plotting** to visualize `sbp` (systolic blood pressure) against other predictors.

```python
import matplotlib.pyplot as plt
import pandas as pd

query = "sbp"

# 0.4: Visualize the distribution of systolic blood pressure
df[query].hist(bins=20)
plt.title(f"Distribution of {query}")
plt.xlabel(query)
plt.ylabel("Frequency")
plt.show()

# 0.5: Explore relationships with continuous predictors
predictors = ["age", "bmi", "sodium", "exercise_minutes", "weight"]

for col in predictors:
    plt.figure()
    plt.scatter(df[col], df[query], alpha=0.7)
    plt.title(f"{query} vs {col}")
    plt.xlabel(col)
    plt.ylabel(query)
    plt.show()

import matplotlib.pyplot as plt
import pandas as pd

colors = {0: "tab:blue", 1: "tab:orange"}

# 0.4: Visualize the distribution of systolic blood pressure
df[query].hist(bins=20)
plt.title(f"Distribution of {query}")
plt.xlabel(query)
plt.ylabel("Frequency")
plt.show()
```
